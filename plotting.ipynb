# experiment plot

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import smooth


# EXP_DIR = 'MountainCar-v0_h=128_seed=42_ep=5000_disc=0.8'
EXP_DIR = '"hidden": 128, "memory": 10000, "env": "MountainCar-v0", "seed": 42, "lr": 0.001, "epsilon": 0.5, "discount": 0.8, "batch_size": 64, "episodes": 3000, "epsilon_cap": 1000, "max_steps": 8000, "target": true'


results = np.load(f'experiments/{EXP_DIR}/ep_durations.npy')


def plot_results(x, N=1, save=False):
    smoothed = smooth(x, N)
    plt.figure(figsize=(20,10))
    plt.plot(smoothed)
    plt.title(f'smoothing N={N}')
    plt.xlabel('Episodes')
    plt.ylabel('Steps')
    if save:
        plt.savefig(f'experiments/{EXP_DIR}/results_smooth={N}.png')
    plt.show()
```

### raw plot

```python
# set smoothing
smoothing_factor = 10


plot_results(results, N=smoothing_factor)
```

### Save the plot

```python
plot_results(results, N=smoothing_factor, save=True)
```
